```python
from urllib.request import urlopen as ureq
from bs4 import BeautifulSoup as soup

url="https://aquahuna.com/collections/tetra"
uclient = ureq(url)

html_content = uclient.read()

uclient.close()
html_soup = soup(html_content, "html.parser")
#html_soup.h1
```

```python
containers = html_soup.findAll("div",{"class":"product grid__item medium-up--one-third small--one-half slide-up-animation animated"})

#print(len(containers))
#print(containers[2])
#print(containers[8])
```

```python
import pandas as pd

itemname_list = []
price_list = []
soldout_list = []


for container in containers:
  try:
    name = container.findAll("div",{"class":"product__title text-center"})
    itemname_list.append(name[0].text.replace('\n',''))
    priced = container.findAll("div",{"class":"product__prices text-center"})
    price = (re.findall("\d+", priced[0].text))
    price_list.append(price[0] + "." + price[1])
    if "Sold Out" in priced[0].text:
      soldout_list.append("Sold Out")
    else: soldout_list.append("In Stock")

  except:
    continue

data = {'Name - Quantity':itemname_list, 'Price':price_list, 'Sold Out or In Stock':soldout_list}
df = pd.DataFrame(data)

f = open("tetra_prices.csv", 'w')
df.to_csv(f)
f.close()

df
```
